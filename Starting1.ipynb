PRE PROCESSING
1. Lower case
2. Tokenisation
3. Removing Special Character
4. Removing Stopwords and Punctuation
5. Stemming

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
import string
string.punctuation
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
ps.stem('Dancing')
```

```python
def text_transform(text):
    text = text.lower()
    text = nltk.word_tokenize(text)

    y=[]
    for i in text:
        if i.isalnum():
            y.append(i)
    text = y[:]
    y.clear()

    for i in text:
        if i not in stopwords.words('english') and i not in string.punctuation:
            y.append(i)

    text = y[:]
    y.clear()

    for i in text :
        y.append(ps.stem(i))


    return " ".join(y)
```

```python
chunk_size = 100  # You can adjust this size depending on your available memory

# Apply the transformation in chunks without reading from or writing to files
for start in range(0, len(df), chunk_size):
    # Define the end of the chunk
    end = start + chunk_size

    # Extract the chunk of the DataFrame
    chunk = df[start:end]

    # Apply the text transformation function to the 'Text' column in the chunk
    df.loc[start:end-1, 'Text'] = chunk['Text'].apply(text_transform)

# Print the transformed DataFrame
print(df)
```
