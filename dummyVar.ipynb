```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics
```

```python
df = pd.read_csv('homepricescat.csv')
df.head()
```

```python
dummies = pd.get_dummies(df.town,dtype=int)
```

```python
merged = pd.concat([df,dummies],axis = 'columns')
```

```python
merged
```

```python
final = merged.drop(['town','west windsor'],axis=1)
final
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
X = final.drop('price',axis = 1)
X
```

```python
y = df.price
y
```

```python
model.fit(X,y)
```

```python
model.predict([[2800,0,1]])
```

```python
model.predict([[3400,0,0]])
```

```python
model.score(X,y) *100
```

```python
car = pd.read_csv('carprices.csv')
car.head()
```

```python
car
```

```python
model_car = pd.get_dummies(car['Car Model'],dtype=int)
model_car
```

```python
merged_car = pd.concat([car,model_car],axis=1)
```

```python
merged_car
```

```python
merged_car = merged_car.drop('Car Model',axis =1)
```

```python
X = merged_car.drop('Sell Price($)',axis =1)
X
```

```python
y = merged_car['Sell Price($)']
y
```

```python
car_regg = LinearRegression()
car_regg.fit(X,y)
```

```python
car_regg.predict([[86000,7,0,1,0]])
```

```python
car_regg.score(X,y)*100
```
