```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics
```

```python
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv')
df.head()
```

```python
sns.lmplot(x ='area',y='price',data = df)
```

```python
plt.xlabel('area')
plt.ylabel('price')
plt.title('Homeprices')
plt.scatter(df.area,df.price,color='red',marker='+')
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df['price'])
```

```python
reg.predict([[3300]])
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
# y = m*x+b
135.78767123*3300+180616.43835616432
```

```python
reg.predict([[5000]])
```

```python
d = pd.read_csv('areas.csv')
d.head()
```

```python
p = reg.predict(d)
```

```python
d['prices'] = p
```

```python
d
```

```python
d.to_csv('predictions.csv')
```

```python
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker="+")
plt.plot(df.area,reg.predict(df[['area']]),color = 'blue')
plt.show()
```

```python
df1 = pd.read_csv('canada_per_capita_income.csv')
df1.head()
```

```python
linear_reg = linear_model.LinearRegression()
linear_reg.fit(df1[['year']],df1[['per capita income (US$)']])
```

```python
linear_reg.predict([[2002]])
```

```python
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df1.year,df1['per capita income (US$)'],color='red',marker="+")
plt.plot(df1.year,linear_reg.predict(df1[['year']]),color='blue')
plt.show()
```
