```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('DailyDelhiClimateTest.csv')
df.head()
```

```python
df.drop(columns = ['humidity','wind_speed','meanpressure'],axis=1,inplace = True)
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['date'],df['meantemp'],color='red')
```

```python
X1 = df.meantemp.ewm(alpha =0.1).mean()
```

```python
X1
```

```python
df['ewma'] = X1
```

```python
df.head()
```

```python
plt.scatter(df['date'],df['meantemp'],color='blue')
plt.plot(df['date'],df['ewma'],color='red')
plt.show()
```
